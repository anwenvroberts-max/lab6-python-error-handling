# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
# 1 - Cataloging: Gather product names
products = []
while True:
    product_name = input("Enter product name to catalog (or type 'done'): ").strip().lower()
    if product_name == 'done':
        if not products:
            print("Catalog is empty. Please add at least one product.")
            continue
        break
    if product_name:
        products.append(product_name)

# 2 - Inventory: Set stock levels (Task 1: Handle non-numeric/negative)
def initialize_inventory(product_list):
    inventory = {}
    print("\n>>> Setting up stock levels:")
    for product in product_list:
        while True:
            try:
                qty = int(input(f"  Number of {product} in stock: "))
                if qty < 0:
                    print("  Error: Please enter a positive number.")
                    continue
                inventory[product] = qty
                break
            except ValueError:
                print("  Error: Please enter a whole number.")
    return inventory

inventory = initialize_inventory(products)

# 3 - Ordering: Get customer items (Task 3: Check existence and stock)
def get_customer_orders(inventory):
    print(f"\nAvailable products: {', '.join(inventory.keys())}")
    orders = set()
    
# This removes spaces AND any accidental double or single quotes
    while True:
        item = input("Enter product name to order (or type 'done' to finish): ").strip().lower().replace('"', '').replace("'", "")
        
        if item == 'done':
            break
        
        # ERROR CHECKING HAPPENS HERE IMMEDIATELY
        if item not in inventory:
            print(f"  -> Wrong item! '{item}' is not in our catalog.")
            print(f"     Reminder, we have: {', '.join(inventory.keys())}")
        elif inventory[item] <= 0:
            print(f"  -> Sorry, '{item}' is currently out of stock.")
        else:
            orders.add(item)
            print(f"  -> Added {item} to your order.")
            
    return orders

customer_orders = get_customer_orders(inventory)

# 4 - Pricing: Get prices (Task 2: Handle non-numeric/negative)
def calculate_total_price(orders):
    if not orders:
        return 0.0
    
    print("\n>>> Enter pricing for ordered items:")
    total = 0.0
    for item in orders:
        while True:
            try:
                price = float(input(f"  Price for {item}: "))
                if price < 0:
                    print("  Error: Price cannot be negative.")
                    continue
                total += price
                break
            except ValueError:
                print("  Error: Please enter a numeric price (e.g., 12.99).")
    return total

total_amount = calculate_total_price(customer_orders)

# 5 - Final Results
def update_inventory(orders, current_inventory):
    # Standard logic: subtract 1 from stock for ordered items
    for item in orders:
        if item in current_inventory:
            current_inventory[item] -= 1
    return current_inventory

inventory = update_inventory(customer_orders, inventory)

print("\n" + "="*30)
print(f"ORDER SUMMARY: {', '.join(customer_orders) if customer_orders else 'No items'}")
print(f"TOTAL AMOUNT DUE: ${total_amount:.2f}")
print("UPDATED STOCK:", inventory)
print("="*30)
````
